```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 날씨 데이터 불러오기
path = 'http://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'

# CSV 파일 읽기
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시' 열을 날짜 형식으로 변환 후, 월 정보 추출
weather['month'] = pd.to_datetime(weather['일시']).dt.month

# 월별 평균 풍속 계산
monthly_wind = [0 for _ in range(12)]  # 12개월 저장할 리스트 생성

for i in range(12):
    monthly_data = weather[weather["month"] == (i + 1)]  # 해당 월 데이터 필터링
    monthly_wind[i] = monthly_data["평균풍속"].mean()  # 평균 풍속 계산

# 막대 그래프 생성
months = np.arange(1, 13)  # 1월부터 12월까지의 월 숫자 생성
plt.bar(months, monthly_wind, color='blue')

# 그래프 라벨 설정
plt.xlabel("Month")
plt.ylabel("Average Wind Speed")
plt.title("Monthly Average Wind Speed")

# 그래프 표시
plt.show()
```
